### EZanchor Quick Start Tutorial Notebook

Welcome to the EZanchor tutorial notebook. Please refer to the main Github README for further guidance.

```python
import ezanchor as eza
```

### Step 1: Instantiate an equipment object

```python
AHU4 = eza.equipment.Equipment(name = "AHU4-stiltMode", 
                               Sds = 1.85, 
                               Ip = 1, 
                               h = 44, 
                               z = 44, 
                               ap = 2.5, 
                               Rp = 2, 
                               omega = 2, 
                               weight = 3500, 
                               CGz = 64, 
                               CGx = 40, 
                               CGy = 85, 
                               load_combo = "LRFD", 
                               use_omega = True)
```

### Step 2: Add equipment footprint

```python
AHU4.add_footprint(xo=0, yo=0, b=60, h=120)
AHU4.add_footprint(xo=60, yo=60, b=60, h=60)
```

```python
fig = eza.plotter.preview(AHU4)
```

### Step 3: Add anchors

Anchors may be added in groups as a rectangular array. Alternatively, they may be added one at a time by specifying their coordinates

```python
AHU4.add_anchor_group(x0=5, y0=5, b=50, h=110, nx=2, ny=3, mode="p")
AHU4.add_anchor_group(x0=65, y0=65, b=50, h=50, nx=2, ny=2, mode="p")
```

```python
fig = eza.plotter.preview(AHU4)
```

```python
AHU4.add_anchor(x=30, y=-5)
AHU4.add_anchor(x=30, y=125)
AHU4.add_anchor(x=90, y=55)
AHU4.add_anchor(x=90, y=125)
```

```python
fig = eza.plotter.preview(AHU4)
```

### Step 4: Solve

```python
AHU4.solve(on_stilt=True)
```

Let's repeat but this time solving using "pivot mode"

```python
AHU5 = eza.equipment.Equipment(name = "AHU5-pivotMode", 
                               Sds = 1.85, 
                               Ip = 1, 
                               h = 44, 
                               z = 44, 
                               ap = 2.5, 
                               Rp = 2, 
                               omega = 2, 
                               weight = 3500, 
                               CGz = 64, 
                               CGx = 40, 
                               CGy = 85, 
                               load_combo = "LRFD", 
                               use_omega = True)
AHU5.add_footprint(xo=0, yo=0, b=60, h=120)
AHU5.add_footprint(xo=60, yo=60, b=60, h=60)
AHU5.add_anchor_group(x0=5, y0=5, b=50, h=110, nx=2, ny=3, mode="p")
AHU5.add_anchor_group(x0=65, y0=65, b=50, h=50, nx=2, ny=2, mode="p")
AHU5.add_anchor(x=30, y=-5)
AHU5.add_anchor(x=30, y=125)
AHU5.add_anchor(x=90, y=55)
AHU5.add_anchor(x=90, y=125)
AHU5.solve(on_stilt=False)
```

### Step 5: Visualization

Compare the two solution schemes

```python
fig3 = eza.plotter.plot_equipment(AHU4)
```

```python
fig4 = eza.plotter.plot_equipment(AHU5)
```

The user may also plot tension and shear demands of a specific anchor

```python
fig5 = eza.plotter.plot_anchor(AHU4, 2)
```

Lastly, the user can export all internal calculation results to csv files for further analysis

```python
AHU4.export_data()
```
